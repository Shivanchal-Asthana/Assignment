```python
#import necessary libraries
import pandas as pd
import io
import time

#import plotty
import warnings
warnings.filterwarnings('ignore')
```

```python
#uploading csv file
from google.colab import files
 
 
uploaded = files.upload()
```

```python
#read the csv file
df1 = pd.read_csv(io.BytesIO(uploaded['Equity Data (1 Jan 2020).csv']))
print(df1)
```

```python
#split the date column into three parts day, month and year
df1[' DATE1']=pd.to_datetime(df1[' DATE1'])
df1['Month']=df1[' DATE1'].apply(lambda x:x.month)
df1['Day']=df1[' DATE1'].apply(lambda x:x.day)
df1['Years']=df1[' DATE1'].apply(lambda x:x.year)
```

```python
#drop the uncessary column
df1.drop(columns=" DATE1", inplace = True)
```

```python
#calculate the percentage change of close price and prev_price 
df1["change"]=(((df1[" CLOSE_PRICE"]-df1[" PREV_CLOSE"])/df1[" PREV_CLOSE"])*100)
```

```python
# Transfer some columns into new dataframe df2
df2 = df1[['SYMBOL','change','Day','Month','Years']]
```

```python
df2
```

```python
#group by whole data based on 'Month'
a = df2.groupby('Month')
```

```python
#To iterate these groups of months one by one
for month, month_df in a:
  print(month)
  print(month_df)
```

```python
#check the maximum performing stock per month
a.max()
```

```python
#check the minimum performing stock per month
a.min()
```
